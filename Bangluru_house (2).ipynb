```python
def convertrange(x):
  if isinstance(x,float):
    return x
  temp=x.split('_')
  if len(temp)==2:
    return float(temp[0])+float(temp[1])/2
  try:
    return float(x)
  except ValueError:
    return None

```

```python
df['total_sqft']=df['total_sqft'].apply(convertrange)
```

```python
#price per squrre fit
df['price_per_squre_fit']=df['price']*100000/df['total_sqft']
```

```python
def removeer_outlier_sqft(df1):
  df1_output=pd.DataFrame()
  for key,subdf1 in df1.groupby('location'):
    m=np.mean(subdf1.price_per_squre_fit)

    st=np.std(subdf1.price_per_squre_fit)
    gen_df1=subdf1[(subdf1.price_per_squre_fit>(m-st))&(subdf1.price_per_squre_fit<=(m+st))]
    df1_output=pd.concat([df1_output,gen_df1],ignore_index=True)
  return  df1_output


df1=removeer_outlier_sqft(df)
df.describe()


```

```python
def bhk_outlier_remover(df1):
  exclude_indices=np.array([])
  for location ,location_df1 in df1.groupby('location'):

    bhk_stats={}
    for bhk,bhk_df1 in location_df1.groupby('BHK'):
      bhk_stats[bhk]={
          'mean':np.mean(bhk_df1.price_per_squre_fit),
          'std':np.std(bhk_df1.price_per_squre_fit),
          'count':bhk_df1.shape[0]

      }
    for bhk,bhk_df1 in location_df1.groupby('BHK'):
      stats=bhk_stats.get(bhk-1)
      if stats and stats['count']>5:
        exclude_indices=np.append(exclude_indices,bhk_df1[bhk_df1.price_per_squre_fit<(stats['mean'])].index.values)
  return df1.drop(exclude_indices,axis='index')


```

```python
df=bhk_outlier_remover(df)
df
```

```python
df.drop(columns=['size','price_per_squre_fit'],inplace=True)
```

```python
X=df.drop(columns=['price'])
y=df['price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=101)
```

```python

#Applying  linear regression
column_trans=make_column_transformer((OneHotEncoder(sparse=False),['location']),
                                     remainder='passthrough')
```

```python
column_trans
```

```python

scaler=StandardScaler()
lr=LinearRegression()
```

```python
lr = LinearRegression(fit_intercept=True)
```

```python
pipe=make_pipeline(column_trans,scaler,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_lr=pipe.predict(X_test)
```

```python
y_pred_lr
```

```python
r2_score(y_test,y_pred_lr)
```

```python
#applying lasso
lasso=Lasso()
```

```python
pipe=make_pipeline(column_trans,scaler,lasso)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_lasso=pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred_lasso)
```

```python
#Applying ridge
ridge=Ridge()
```

```python
pipe=make_pipeline(column_trans,scaler,ridge)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_ridge=pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred_ridge)
```

```python
print('NO Regularization:', r2_score(y_test,y_pred_lr))
print('Lasso:',r2_score(y_test,y_pred_lasso))
print('Ridge:',r2_score(y_test,y_pred_ridge))
```
